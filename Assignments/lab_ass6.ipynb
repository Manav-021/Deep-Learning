```python
import numpy as np;
import pandas as pd;
```

```python
X=np.linspace(0,10,50).reshape(-1,1)
```

```python
#target should increase non linearly
noise=np.random.normal(0,0.2,size=(50,1))
y=np.log(X+1)+noise
y.shape
```

```python
n_layers=3
```

```python
weights_input=np.random.uniform(-1,1,size=(1,n_layers));
bias_hidden=np.zeros((1,n_layers));
weights_hidden=np.random.uniform(-1,1,size=(n_layers,1));
bias_output=np.zeros((1,1));

print(weights_input.shape)
print(weights_hidden.shape)
```

```python
def Activation(x):
    return np.maximum(0,x);
```

```python
def activation_slope(z):
    return (z > 0).astype(float)
```

```python
def forward_pass(x,weights_input,weights_hidden,bias_hidden,bias_output):
    z=x@weights_input+bias_hidden;
    h=Activation(z);
    y_hat=h@weights_hidden+bias_output;

    return z,h,y_hat
```

```python
def error(y,y_hat):
    return y-y_hat;
```

```python
def lossfn(y, y_hat):
  return np.mean((y - y_hat)**2)
```

```python
learning_rate=0.001;

for epoch in range(1000):
  z1 = X @ weights_input+ bias_hidden
  h = Activation(z1)
  y_hat = h @ weights_hidden + bias_output


  error = y_hat - y
  loss = np.mean(error **2)

  dL_dy =2 * error /len(X)
  dL_dW2 = h.T @ dL_dy
  dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

  dL_dh = dL_dy @ weights_hidden.T
  dL_dz1 = dL_dh * activation_slope(z1)

  dL_dW1 = X.T @ dL_dz1
  dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

  weights_input -= learning_rate * dL_dW1
  bias_hidden -= learning_rate * dL_db1
  weights_hidden -= learning_rate * dL_dW2
  bias_output -= learning_rate * dL_db2

  if epoch % 100 == 0:
      print(f"Epoch {epoch:04d} | Loss: {loss:.6f}")


for i in range(5):
  print(f"X: {X[i][0]:.2f} | True: {y[i][0]:.4f} | Pred: {y_hat[i][0]:.4f}")
```
